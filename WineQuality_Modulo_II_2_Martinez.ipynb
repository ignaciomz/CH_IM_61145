# GLOSARIO

**Fixed Acidity** (Acidez Fija): Representa la cantidad total de 치cidos no vol치tiles presentes en el vino, como el 치cido tart치rico y el 치cido m치lico. Contribuye a la estructura y estabilidad del vino.

**Volatile Acidity** (Acidez Vol치til): Indica la concentraci칩n de 치cidos vol치tiles, principalmente el 치cido ac칠tico, en el vino. Altos niveles de acidez vol치til pueden dar lugar a aromas y sabores desagradables, como el vinagre.

**Citric Acid** (츼cido C칤trico): Es un 치cido org치nico que se encuentra de forma natural en muchas frutas c칤tricas. En el vino, puede contribuir a la acidez total y proporcionar un sabor fresco y c칤trico.

**Residual Sugar** (Az칰car Residual): Es la cantidad de az칰car que permanece sin fermentar en el vino despu칠s del proceso de fermentaci칩n. Contribuye al cuerpo, dulzura y equilibrio del vino.

**Chlorides** (Cloruros): Representa la concentraci칩n de cloruros en el vino, que puede provenir tanto de la uva como del proceso de vinificaci칩n. Puede afectar el sabor y la percepci칩n de la acidez del vino.

**Free Sulfur Dioxide** (Di칩xido de Azufre Libre): Es una forma de di칩xido de azufre (SO2) que act칰a como antioxidante y conservante en el vino. Ayuda a prevenir la oxidaci칩n y el crecimiento de microorganismos no deseados.

**Total Sulfur Dioxide** (Di칩xido de Azufre Total): Representa la cantidad total de di칩xido de azufre (libre m치s unido) presente en el vino. Se utiliza como medida de la estabilidad y conservaci칩n del vino.

**Density**(Densidad): Indica la densidad del vino, generalmente medida en g/ml. Puede estar relacionada con el contenido de az칰car y alcohol del vino, as칤 como con su cuerpo y viscosidad.

**pH**: Es una medida de la acidez o alcalinidad del vino en una escala de 0 a 14, donde 7 es neutro. Un pH m치s bajo indica mayor acidez, mientras que un pH m치s alto indica menor acidez.

**Sulphates** (Sulfatos): Representa la concentraci칩n de sulfatos en el vino, que pueden provenir de aditivos como el metabisulfito de potasio. Se utilizan como conservantes y antioxidantes en el vino.

**Alcohol**: Indica el contenido de alcohol en volumen (% ABV) del vino. Afecta el cuerpo, el sabor y la sensaci칩n de calor en el vino.

**Quality** (Calidad): Es una medida subjetiva de la calidad percibida del vino, generalmente evaluada por catadores o expertos enol칩gicos. Puede basarse en m칰ltiples factores, incluidos el aroma, sabor, textura y equilibrio del vino.

```python
# Probemos remover outliers.

Q1 = df['citric acid'].quantile(0.25)
Q3 = df['citric acid'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_ac_eliminado = df[(df['citric acid'] >= lower_bound) & (df['citric acid'] <= upper_bound)]

print("\nValores sin outliers:")
df.describe().round(2)
```

```python
# Comenzamos la prueba gr치fica.

plt.figure(figsize=(10, 6))
plt.gca().set_facecolor('lightgrey')
sns.histplot(data=df_ac_eliminado, x='citric acid', hue='quality', palette='Purples', multiple='stack', bins=10)
plt.xlabel('Citric Acid')
plt.ylabel('Frecuencia')
plt.title('Cantidad de datos seg칰n Citric Acid y Densidad')
plt.show()
```
